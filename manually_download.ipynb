```python
import pickle as pkl
import json

def convert_dict_to_str(d):
    # 如果输入是字典，则遍历每个键值对
    if isinstance(d, dict):
        return {str(k): convert_dict_to_str(v) for k, v in d.items()}
    # 如果输入是列表或元组，递归处理其元素
    elif isinstance(d, list):
        return [convert_dict_to_str(i) for i in d]
    elif isinstance(d, tuple):
        return tuple(convert_dict_to_str(i) for i in d)
    # 如果是其他数据类型，则直接转换为字符串
    else:
        return str(d)

with open("/home/ksas/Public/datasets/cub_concept_bank/run_info-pcbm_cub__resnet18_cub__cub_resnet18_cub_0__lam:0.0002__alpha:0.99__seed:42.pkl", "rb") as input_stream:
    run_info = pkl.load(input_stream)
    print(json.dumps(convert_dict_to_str(run_info), indent=4))
    
with open("/home/ksas/Public/datasets/cifar10_concept_bank/run_info-pcbm_cifar10__clip:RN50__multimodal_concept_clip:RN50_cifar10_recurse:1__lam:0.0002__alpha:0.99__seed:42.pkl", "rb") as input_stream:
    run_info = pkl.load(input_stream)
    print(json.dumps(convert_dict_to_str(run_info), indent=4))
```
